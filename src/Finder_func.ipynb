```python
def separate_code(tokens):
    code_chunks = [[]]
    for token in tokens:
        code_chunks[-1].append(token)
        if token == 'dot':
            code_chunks.append([])
    code_chunks = [chunk for chunk in code_chunks if chunk]
    return code_chunks
```
